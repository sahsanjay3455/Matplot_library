```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,5)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")

plt.show()
```

```python
# changing the line aethetics
plt.plot(x,y,color='g',linestyle=':',linewidth=3)
plt.title("line plot ")
plt.xlabel("this is x-label")
plt.ylabel("this is y-label")
plt.show()
```

```python
# using the grid attribute in the graph 


plt.plot(x,y,color='g',linestyle=':',linewidth=3)
plt.title("line plot ")
plt.xlabel("this is x-label")
plt.ylabel("this is y-label")

plt.grid(True)
plt.show()
```

```python
z=x*3
z
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=3)
plt.plot(x,z,color='b',linestyle='-.',linewidth=4)

plt.title("line plot ")
plt.xlabel("this is x-label")
plt.ylabel("this is y-label")

plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)

plt.plot(x,y,color='g',linestyle=':',linewidth=3)
plt.plot(x,z,color='b',linestyle='-.',linewidth=4)

plt.title("line plot ")
plt.xlabel("this is x-label")
plt.ylabel("this is y-label")

plt.grid(True)


plt.subplot(1,2,2)

plt.plot(x,y,color='y',linestyle=':',linewidth=3)
plt.plot(x,z,color='r',linestyle='-.',linewidth=4)

plt.title("line plot ")
plt.xlabel("this is x-label")
plt.ylabel("this is y-label")

plt.grid(True)
plt.show()
```

```python
# Bar graph
from matplotlib import pyplot as plt
student={"sanjay":88,"ajay":98,"manjay":50,"ranjan":78,"sunil":77}
names=list(student.keys())
values=list(student.values())
plt.title("Bar-graph")
plt.xlabel("Name of students")
plt.ylabel("Number of marks")
plt.bar(names,values)
plt.show()
```

```python
# horizontal Bar graph
from matplotlib import pyplot as plt
student={"sanjay":88,"ajay":98,"manjay":50,"ranjan":78,"sunil":77}
names=list(student.keys())
values=list(student.values())
plt.title("Bar-graph")
plt.xlabel("Name of students")
plt.ylabel("Number of marks")
plt.barh(names,values,color='g')
plt.show()
```

```python
# scatter plot
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(5,55,5)
x
```

```python
y=np.random.randint(0,10,10)
y
```

```python
plt.title("scatter plot")
plt.xlabel("frequency")
plt.ylabel("number")
plt.scatter(x,y,marker="*",c='g',s=100)
plt.show()
```

```python
# histogram :it is used for numberical values
# it is used to represent the count the data 

import numpy as np
from matplotlib import pyplot as plt
n1=np.random.randint(1,10,20)
n1
plt.hist(n1,color='y')
plt.title("Histogram")
plt.xlabel("frequency")
plt.ylabel("number")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
n1=np.random.randint(1,10,20)
n1
plt.hist(n1,color='b',bins=4)
plt.title("Histogram")
plt.xlabel("frequency")
plt.ylabel("number")
plt.show()
```

```python
# box-plot
from matplotlib import pyplot as plt
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,6,5,4]
data=list([one,two ,three])

plt.title("Box-plot")
plt.xlabel("frequency")
plt.ylabel("number")
plt.boxplot(data)
colors=['red','blue','yellow']
plt.show()

```

```python
# pie-chart
# it contain quantity and labels
fruits=['mango','apple','orange','banana','gauava']
quantity=[10,5,15,20,8]
plt.pie(quantity,labels=fruits)

plt.show()
```
